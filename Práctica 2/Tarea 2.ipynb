#### Paquetes necesarios

```python
import cv2  
import matplotlib.pyplot as plt
```

#### Selección de imagen desde archivo y conversión a escala de grises

```python
# Lee imagen de archivo
img = cv2.imread('Images/grecia.jpg') 

# Conversión de la imagen a escala de grises desde BGR
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Muestra la imagen
plt.figure()
plt.axis("off")
plt.imshow(gris, cmap="gray")
plt.show()
```

#### Uso del operador de Sobel 

```python
# Filtro Gaussiano para suavizar la imagen original
ggris = cv2.GaussianBlur(gris, (9, 9), 0)

# Calcula en ambas direcciones (horizontal y vertical)
sobelx = cv2.Sobel(ggris, cv2.CV_64F, 1, 0)  # x
sobely = cv2.Sobel(ggris, cv2.CV_64F, 0, 1)  # y

#Combina ambos resultados
sobel = cv2.add(sobelx, sobely)

# Convierte a escala manejable (Valores de 0 a 255)
sobelx_scaled = cv2.convertScaleAbs(sobelx)
sobely_scaled = cv2.convertScaleAbs(sobely)
sobel_scaled = cv2.convertScaleAbs(sobel)

# Lista de títulos para los subplots
titles = ['Verticales', 'Horizontales', 'Combinada']

# Lista de imágenes para mostrar en los subplots
images = [sobelx, sobely, sobel, sobelx_scaled, sobely_scaled, sobel_scaled]

# Crear una figura
fig, axs = plt.subplots(2, 3, figsize=(10, 6))

# Recorrer las filas y columnas de subplots
for i in range(2):
    for j in range(3):
        index = i * 3 + j  # Calcular el índice de la imagen y el título
        axs[i, j].imshow(images[index], cmap='gray')  # Mostrar la imagen
        axs[i, j].set_title(titles[j])  # Establecer el título
        axs[i, j].axis('off')  # Apagar los ejes

plt.tight_layout()  # Ajustar el espaciado automáticamente
plt.show()
```

#### Contenido de la imagen escalada y sin escalar

```python
# Contenido de imagen con operador Sobel sin escalar
# Operador Sobel en x (Detección de bordes verticales)
print("Matriz sobel_x sin escalar: \n{}".format(sobelx))

# Contenido de imagen con operador Sobel escalada
# Operador Sobel en x (Detección de bordes verticales)
print("\nMatriz sobel_x escalada: \n{}".format(sobelx_scaled))
```
