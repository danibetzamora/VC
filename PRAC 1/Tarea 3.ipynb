#### Paquetes necesarios

```python
import cv2  
import numpy as np
import matplotlib.pyplot as plt
```

#### Inicializaciones prueba de dibujo

```python
# Dimensiones de la imagen
ancho = 800
alto = 800
```

#### Prueba con funciones de dibujo de OpenCV

```python
# Crear una matriz de ceros (imagen en blanco) con dimensiones (ancho, alto, 3) y tipo de datos uint8
img_drawn = np.zeros((ancho, alto, 3), dtype=np.uint8)

# Dibujar un círculo rojo en la parte superior izquierda
cv2.circle(img_drawn, (ancho - 400, 120), 50, (0, 0, 255), -1)

# Dibujar una elipse azul en la parte inferior izquierda
cv2.ellipse(img_drawn, (256, alto - 100), (100, 50), 0, 0, 180, (255, 0, 0), -1)

# Dibujar un círculo verde con un borde grueso en la parte superior derecha
cv2.circle(img_drawn, (ancho - 70, 350), 50, (0, 255, 0), 20)

# Dibujar un círculo rojo grande en el centro
cv2.circle(img_drawn, (400, 400), 300, (255, 0, 0), 5)

# Dibujar un rectángulo verde con un borde grueso
cv2.rectangle(img_drawn, (500, 300), (350, 400), (0, 255, 0), 40)

# Dibujar una línea roja diagonal
cv2.line(img_drawn, (int(150/3), 0), (int(550/2), 150), (0, 0, 255), 8)

# Mostrar la imagen en una ventana utilizando Matplotlib
plt.imshow(img_drawn, cmap='gray')
plt.show()
```

#### Inicializaciones imagen Mondrian 

```python
# Dimensiones
ancho = 200
alto = 200
```

#### Imagen estilo Mondrian con funciones de dibujo de OpenCV

```python
# Crear imagen con tres canales
color_img = np.zeros((alto, ancho, 3), dtype = np.uint8)

# Poner imagen a blanco cada vez que se ejecute el código
color_img[:,:,0] = 255
color_img[:,:,1] = 255 
color_img[:,:,2] = 255

# Líneas negras
cv2.line(color_img, (50,0), (50, 200), (0,0,0), 4)
cv2.line(color_img, (0,60), (48, 60), (0,0,0), 5)
cv2.line(color_img, (0,150), (200, 150), (0,0,0), 4)
cv2.line(color_img, (175,150), (175, 200), (0,0,0), 3)
cv2.line(color_img, (175,175), (200, 175), (0,0,0), 4)
cv2.line(color_img, (50,100), (200, 100), (0,0,0), 4)
cv2.line(color_img, (100,0), (100, 100), (0,0,0), 4)
cv2.line(color_img, (150,0), (150, 100), (0,0,0), 4)
cv2.line(color_img, (150,50), (200, 50), (0,0,0), 4)
cv2.line(color_img, (125,100), (125, 150), (0,0,0), 4)
cv2.line(color_img, (100,150), (100, 200), (0,0,0), 4)

# Rectángulo rojo superior centro
cv2.rectangle(color_img,(53,0),(97,97),(255,0,0),-1)

# Rectángulo azul inferior izquierda
cv2.rectangle(color_img,(0,153),(47,200),(0,0,255),-1)

# Rectángulo azul inferior derecha
cv2.rectangle(color_img,(103,153),(172,200),(0,0,255),-1)

# Rectángulo amarillo inferior derecha
cv2.rectangle(color_img,(178,178),(200,200),(255,255,0),-1)

# Rectángulo azul superior derecha
cv2.rectangle(color_img,(153,0),(200,47),(0,0,255),-1)

# Rectángulo amarillo superior izquierda
cv2.rectangle(color_img,(0,0),(47,56),(255,255,0),-1)

# Rectángulo amarillo superior derecha
cv2.rectangle(color_img,(153,53),(200,97),(255,255,0),-1)

# Rectángulo central derecha
cv2.rectangle(color_img,(128,103),(200,147),(255,0,0),-1)

# Rectángulo amarillo central
cv2.rectangle(color_img,(53,103),(122,147),(255,255,0),-1)

# Visualización de la imagen
plt.imshow(color_img) 
plt.show()
```
