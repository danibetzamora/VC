#### Paquetes necesarios

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Inicializaciones

```python
# Dimensiones de la imagen
ancho = 200
alto = 200
```

#### Creación imagen con tres canales

```python
# Crear imagen con tres canales
color_img = np.zeros((alto, ancho, 3), dtype = np.uint8)

# Dimensiones de la imagen
print(color_img.shape)

# Visualización de imagen base
plt.imshow(color_img)
plt.show()
```

#### Creación imagen estilo Mondrian
