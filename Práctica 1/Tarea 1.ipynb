#### Paquetes Necesarios

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Inicializaciones

```python
# Dimensiones del tablero de ajedrez
ancho = 800
alto = 800
```

#### Creación de imagen

```python
# Crea una imagen de un único plano (Al rellenar la matriz con 0, será negra)
tablero = np.zeros((ancho, alto,1), dtype=np.uint8)

# Muestra las dimensiones del tablero
print(tablero.shape)
```

#### Visualización del tablero base

```python
# Muestra la imagen con matplotlib
plt.imshow(tablero, cmap='gray')
plt.show()
```

#### Creación del tablero de ajedrez

```python
# Modifica un par de zonas rectangulares de la imagen para crear el tablero
tablero[:] = 0
variable = 0

for fila in range(8):
    for columna in range(8):
        if columna % 2 == variable:
            tablero[fila*100:(fila+1)*100,columna*100:(columna+1)*100,0] = 255
    if fila % 2 == 0:
        variable = 1
    else:
        variable = 0

plt.imshow(tablero, cmap='gray')
plt.show()
```
